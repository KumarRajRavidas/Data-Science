# Null value Anaysis

```python
missing_val_count_by_column = train.isnull().sum()
print(missing_val_count_by_column[missing_val_count_by_column > 0])
```

```python
null_value= train.isnull().sum().sort_values(ascending=False) # to check the null value in data
percentage = (train.isnull().sum()/train.isnull().count()).sort_values(ascending=False)
missing_value_data = pd.concat([null_value, percentage], axis=1, keys=['Total_Null', 'Percentage'])
print(missing_value_data)

null_value_test= test.isnull().sum().sort_values(ascending=False) # to check the null value in data
percentage_test = (test.isnull().sum()/test.isnull().count()).sort_values(ascending=False)
missing_value_data_test= pd.concat([null_value_test, percentage_test], axis=1, keys=['Total_Null_test', 'Percentage_test'])
print(missing_value_data_test)
```

```python
# print(train.isna().sum().sort_values(ascending=False)) # to check the NAN value in data
data = train.drop((missing_value_data[missing_value_data["Percentage"] > 0.04]).index,1) 
data.isnull().sum().sort_values(ascending=False)
```

```python
data_test = test.drop((missing_value_data_test[missing_value_data_test["Percentage_test"] > 0.02]).index,1) 
data_test.isnull().sum().sort_values(ascending=False)
```
