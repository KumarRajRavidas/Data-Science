```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
file_path = 'train.csv'
home_data = pd.read_csv(file_path)
y = home_data.SalePrice

features= ['MSSubClass','LotArea','OverallQual','OverallCond','YearBuilt','YearRemodAdd','BsmtFinSF1',
 'BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','1stFlrSF','2ndFlrSF','LowQualFinSF','GrLivArea','BsmtFullBath',
'BsmtHalfBath','FullBath','HalfBath','BedroomAbvGr','KitchenAbvGr','TotRmsAbvGrd','Fireplaces',
'GarageCars','GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch',
'PoolArea','MiscVal','MoSold','YrSold']

X = home_data[features]
X.head()
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1) # Split into validation and training data
```

```python
# Define a random forest model
# ranf_model = RandomForestRegressor(random_state=1, n_estimators=700)
# ranf_model.fit(train_X, train_y)
# ranf_val_predictions = ranf_model.predict(val_X)

gbm_model = GradientBoostingRegressor(random_state=1, n_estimators=200)
gbm_model.fit(train_X, train_y)
gbm_val_predictions = gbm_model.predict(val_X)
gbm_val_mae = mean_absolute_error(gbm_val_predictions, val_y)
gbm_val_rmse = np.sqrt(mean_absolute_error(gbm_val_predictions, val_y))
print(gbm_val_mae)
print(gbm_val_rmse)
```

```python
# ranf_model_data = RandomForestRegressor(random_state=1, n_estimators=700)
# fit ranf_model_data on all data from the training data
# ranf_model_data.fit(X, y)

gbm_model_on_full_data = GradientBoostingRegressor(random_state=1, n_estimators=200)
gbm_model_on_full_data.fit(train_X, train_y)
```

```python
test_data_path = 'test.csv'
test_data = pd.read_csv(test_data_path)
test_data = test_data.fillna(-1)
test_X = test_data[features]

test_preds = gbm_model_on_full_data.predict(test_X)
print(test_preds)
```

```python
output = pd.DataFrame({'Id': test_data['Id'],'SalePrice': test_preds})
output.to_csv('Pridiction_submission.csv', index=False)
output
```
